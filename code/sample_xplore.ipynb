```python
import os
import mne
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = mne.datasets.sample.data_path()
raw_fname = os.path.join(data_path, "MEG", "sample", "sample_audvis_raw.fif")
raw = mne.io.read_raw_fif(raw_fname, preload=True)
raw.pick_types(meg=False, eeg=True)
data, times = raw[:1, :]  # first EEG channel
sr = raw.info["sfreq"]
index = pd.to_datetime(times, unit="s")
eeg_series = pd.Series(np.ravel(data), index=index)
```

```python
data, times = raw[0, :10]#first 5 channels, first 1000 time points
print(data.shape)
# print(data)
# print(times)

channel = raw.ch_names[0]


values = data.flatten()


# Create one figure only
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(times, values, marker='o')
ax.set_title(f"First {len(times)} samples from {channel} channel")
ax.set_xlabel("Time (s)")
ax.set_ylabel("EEG values (V)")
ax.grid(True)

plt.show()


```
